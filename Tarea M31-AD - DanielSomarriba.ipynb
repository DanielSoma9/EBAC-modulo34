```python
import pandas as pd
import pytz

df = pd.read_csv("fifa_eda.csv")

# Lambda
df['ZonaHoraria'] = df['Nationality'].apply(
    lambda x: pytz.country_timezones[x][0] if x in pytz.country_timezones else ''
)

#Resultados
print(df[['Nationality','ZonaHoraria']].head(20))
```

```python
# Lambda para 'PremioBalonDeOro'
df['BalonDeOro'] = df['Potential'].apply(lambda x: 'Candidato' if x > 85 else 'Next Year')

# Ver resultados
print(df[['Name','Potential','BalonDeOro']].head(20))
```
